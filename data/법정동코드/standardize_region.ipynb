```python
import pandas as pd
from urllib.parse import quote
```

```python
# 한글 파일명을 URL 인코딩
base_url = "https://raw.githubusercontent.com/Zoe3399/1_team_project/main/data/"
file_name = "법정동코드.csv"
encoded_file_name = quote(file_name)

# 최종 URL 조합
url = base_url + encoded_file_name

# CSV 불러오기
df = pd.read_csv(url, encoding="utf-8")
df.head()
```

```python
# 셀 3: 폐지된 항목 제거 및 분리
df_exist = df[df['폐지여부'] == '존재'].copy()

split_names = df_exist['법정동명'].str.split(' ', n=2, expand=True)
split_names.columns = ['시도명', '시군구명', '읍면동명']
for col in ['시도명', '시군구명', '읍면동명']:
    split_names[col] = split_names[col].fillna('').str.strip()

df_exist['시도명'] = split_names['시도명']
df_exist['시군구명'] = split_names['시군구명']
df_exist['읍면동명'] = split_names['읍면동명']
```

```python
# 셀 4: 지역명 생성 및 확인
df_exist['지역명'] = (
    df_exist['시도명'] + ' ' +
    df_exist['시군구명'] + ' ' +
    df_exist['읍면동명']
).str.strip()

df_cleaned = df_exist[['법정동코드', '시도명', '시군구명', '읍면동명', '지역명']]
df_cleaned.head()
```
