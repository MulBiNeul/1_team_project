```python
import pandas as pd
```

```python
df_2020 = pd.read_excel('../data/시도_시군구별_시간대별_노인_교통사고/시도_시군구별_시간대별_노인_교통사고_2020.xls', header=[0, 1])
df_2020.head()
```

```python
df_2021 = pd.read_excel('../data/시도_시군구별_시간대별_노인_교통사고/시도_시군구별_시간대별_노인_교통사고_2021.xls', header=[0, 1])
df_2021.head()
```

```python
df_2022 = pd.read_excel('../data/시도_시군구별_시간대별_노인_교통사고/시도_시군구별_시간대별_노인_교통사고_2022.xls', header=[0, 1])
df_2022.head()
```

```python
df_2023 = pd.read_excel('../data/시도_시군구별_시간대별_노인_교통사고/시도_시군구별_시간대별_노인_교통사고_2023.xls', header=[0, 1])
df_2023.head()
```

```python
df_2024 = pd.read_excel('../data/시도_시군구별_시간대별_노인_교통사고/시도_시군구별_시간대별_노인_교통사고_2024.xls', header=[0, 1])
df_2024.head()
```

```python
# 공통 3개 컬럼 추출
common_cols = [('시도', '시도'), ('시군구', '시군구'), ('기준년도', '기준년도')]

# 시간대 컬럼만 추출
df_2020_times = df_2020.drop(columns=common_cols)
df_2021_times = df_2021.drop(columns=common_cols)
df_2022_times = df_2022.drop(columns=common_cols)
df_2023_times = df_2023.drop(columns=common_cols)
df_2024_times = df_2024.drop(columns=common_cols)

# 병합 처리
df_merged = pd.concat([df_2020[common_cols], df_2020_times, df_2021_times, df_2022_times, df_2023_times, df_2024_times], axis=1)
df_merged.head()
```

```python
### 더 보기 좋게 정렬 ###

common_cols = [('시도', '시도'), ('시군구', '시군구'), ('기준년도', '기준년도')]

# 시간대 순서 고정
time_order = [
    '0시~2시', '2시~4시', '4시~6시', '6시~8시', '8시~10시',
    '10시~12시', '12시~14시', '14시~16시', '16시~18시',
    '18시~20시', '20시~22시', '22시~24시'
]

# '합계' 컬럼 먼저 연도 순 정렬
total_cols = sorted(
    [col for col in df_merged.columns if col[1] == '합계'],
    key=lambda x: int(x[0])
)

# 시간대 컬럼 정렬: 연도 + 시간
time_cols = [
    col for col in df_merged.columns
    if col[1] in time_order and col not in common_cols
]

# 연도별 시간대 순서대로 정렬
time_cols_sorted = sorted(
    time_cols,
    key=lambda x: (int(x[0]), time_order.index(x[1]))
)

final_cols = common_cols + total_cols + time_cols_sorted    # 최종 column 순서 조립
df_time = df_merged[final_cols]                        # 재정렬 적용

# 숫자형(float)인 컬럼만 선택하여 int로 변환
for col in df_time.columns:
    if pd.api.types.is_numeric_dtype(df_time[col]):
        df_time[col] = df_time[col].astype('Int64')

df_time
```
