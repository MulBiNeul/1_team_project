```python
### 더 보기 좋게 정렬 ###

common_cols = [('시도', '시도'), ('시군구', '시군구'), ('기준년도', '기준년도')]

# 시간대 순서 고정
time_order = [
    '0시~2시', '2시~4시', '4시~6시', '6시~8시', '8시~10시',
    '10시~12시', '12시~14시', '14시~16시', '16시~18시',
    '18시~20시', '20시~22시', '22시~24시'
]

# '합계' 컬럼 먼저 연도 순 정렬
total_cols = sorted(
    [col for col in df_merged.columns if col[1] == '합계'],
    key=lambda x: int(x[0])
)

# 시간대 컬럼 정렬: 연도 + 시간
time_cols = [
    col for col in df_merged.columns
    if col[1] in time_order and col not in common_cols
]

# 연도별 시간대 순서대로 정렬
time_cols_sorted = sorted(
    time_cols,
    key=lambda x: (int(x[0]), time_order.index(x[1]))
)

final_cols = common_cols + total_cols + time_cols_sorted    # 최종 column 순서 조립
df_time = df_merged[final_cols]                             # 재정렬 적용

# 숫자형(float)인 컬럼만 선택하여 int로 변환
for col in df_time.columns:
    if pd.api.types.is_numeric_dtype(df_time[col]):
        df_time[col] = df_time[col].astype('Int64')

df_time = df_time.replace('-', 0)                           # 셀에 0이 -로 표시된 것 0으로 변경
df_time

# xlsx 파일로 저장
# df_time.to_excel('시도_시군구_시간대별_노인교통사고_전처리ver.xlsx', index=True)
```
