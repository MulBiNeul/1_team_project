```python
import pandas as pd
from IPython.display import HTML
```

```python
df_casualties_2020 = pd.read_excel('../../data/raw_data/연령층별_월별_사상자/연령층별_월별_사상자_2020.xls', header=[0, 1])
df_casualties_2021 = pd.read_excel('../../data/raw_data/연령층별_월별_사상자/연령층별_월별_사상자_2021.xls', header=[0, 1])
df_casualties_2022 = pd.read_excel('../../data/raw_data/연령층별_월별_사상자/연령층별_월별_사상자_2022.xls', header=[0, 1])
df_casualties_2023 = pd.read_excel('../../data/raw_data/연령층별_월별_사상자/연령층별_월별_사상자_2023.xls', header=[0, 1])
df_casualties_2024 = pd.read_excel('../../data/raw_data/연령층별_월별_사상자/연령층별_월별_사상자_2024.xls', header=[0, 1])
```

```python
df_casualties_2022.columns
```

```python
# 공통 3개 컬럼 추출
common_cols = [( '연령 ', '연령 '), ('사고년도', '사고년도')]

# 시간대 컬럼만 추출
df_casualties_2020_extracted = df_casualties_2020.drop(columns=common_cols)
df_casualties_2021_extracted = df_casualties_2021.drop(columns=common_cols)
df_casualties_2022_extracted = df_casualties_2022.drop(columns=common_cols)
df_casualties_2023_extracted = df_casualties_2023.drop(columns=common_cols)
df_casualties_2024_extracted = df_casualties_2024.drop(columns=common_cols)

# merge
df_casualties = pd.concat(
    [df_casualties_2020[common_cols],
     df_casualties_2020_extracted,
     df_casualties_2021_extracted,
     df_casualties_2022_extracted,
     df_casualties_2023_extracted,
     df_casualties_2024_extracted],
     axis=1)
df_casualties
# HTML(df_casualties.head(5).to_html())
```

```python
### 더 보기 좋게 정렬 ###
common_cols = [( '연령 ', '연령 '), ('사고년도', '사고년도')]

# 시간대 순서 고정
time_order = [
    '01월', '02월', '03월', '04월',
    '05월', '06월', '07월', '08월',
    '09월', '10월', '11월', '12월'
]

# '합계' 컬럼 먼저 연도 순 정렬
total_cols = sorted(
    [col for col in df_casualties.columns if col[1] == '합계'],
    key=lambda x: int(x[0])
)

# 시간대 컬럼 정렬: 연도 + 시간
time_cols = [
    col for col in df_casualties.columns
    if col[1] in time_order and col not in common_cols
]

# 연도별 시간대 순서대로 정렬
time_cols_sorted = sorted(
    time_cols,
    key=lambda x: (int(x[0]), time_order.index(x[1]))
)

final_cols = common_cols + total_cols + time_cols_sorted    # 최종 column 순서 조립
df_casualties = df_casualties[final_cols]                             # 재정렬 적용
df_casualties

# xlsx 파일로 저장
# df_casualties.to_excel('../../data/preprocessed_data/연령층별_월별_사상자_전처리ver.xlsx')  # 자동 인덱스 생성으로 인하여 생긴 공백 행과 인덱스 열은 직접 제거해야 함
```
